## Analyzing the Bengali Handwritten-Character Classifiaction Model

In this section we've tried to create some graphical figures of the trained model to get some insights about the classification model. At first we have plotted two graphs one is based on the Loss function and another one is on accuracy of the train & validate dataset. Secondly, we have plotted both loss & accuracy variables with respect to the epochs. And at last we have plotted the Confusion Matrix of the validation dataset.  

```python
# Importing all the libarires required...
import json
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.metrics import confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
```

#### Plotting the Model Loss graph; loss v.s. epochs

```python
# Loading the loss function files earlier that stored the model statistics
with open("loss_scores","r") as fp:
    loss = json.load(fp)
    fp.close()

with open("val_loss_scores","r") as fp:
    val_loss= json.load(fp)
    fp.close()
```

```python
# Plotting the 'Model Loss' graph

fig, ax = plt.subplots()

ax.plot(loss, color = 'blue')
ax.plot(val_loss, color = 'red')
ax.set_xlabel("Epochs")
ax.set_ylabel("Loss")
ax.legend(['Loss', 'val_loss'], loc = 'upper right')

plt.title("Model Loss")
plt.show()

# Saving the file
fig.savefig("model loss")
```

#### Plotting the Accuracy graph; accuracy v.s epochs

```python
# Loading the accuracy variables files earlier that stored the model statistics
with open("accuracy_scores","r") as fp:
    accuracy = json.load(fp)
    fp.close()

with open("val_accuracy_scores","r") as fp:
    val_accuracy = json.load(fp)
    fp.close()
```

```python
# Plotting the 'Accuracy' graph

fig, ax = plt.subplots()

ax.plot(accuracy, color = 'blue')
ax.plot(val_accuracy, color = 'red')
ax.set_xlabel("Epochs")
ax.set_ylabel("Accuracy")
ax.legend(['accuracy', 'val_accuracy'], loc = 'lower right')

plt.title("Model Accuracy")
plt.show()

# Saving the file
fig.savefig("model accuracy")
```

#### Change of accuracy and error rate (loss function) with each passing epoch

```python
epochs = range(0,5)

fig, ax1 = plt.subplots()
ax1.plot(epochs, accuracy, color = "blue")

ax2 = ax1.twinx()
ax2.plot(epochs, loss, color = "green")

ax1.set_ylabel("Accuracy")
ax2.set_ylabel("Error")

plt.title("Change of accuracy and error rate with epoch")
plt.show()

# Saving the file
fig.savefig("accuracy vs error")
```

#### Plotting the Confusion matrix for the validation dataset

```python
# Loading the model predictiosn and true values
with open("val_predictions","r") as fp:
    val_predictions = json.load(fp)
    fp.close()

with open("val_true","r") as fp:
    val_true = json.load(fp)
    fp.close()
```

```python
# Creating the matrix with the previously loaded variables
mat = confusion_matrix(val_predictions, val_true)

# Plotting the recently created matrix
fig, ax = plot_confusion_matrix(conf_mat = mat, 
                      figsize = (100,100),
                      show_absolute = True, 
                      show_normed = True,
                      colorbar = True)
ax.set_xlabel("Validation Predictions")
ax.set_ylabel("Validation True Values")

plt.title("Confusion Matrix")
plt.show()

# Saving the file
fig.savefig("confusion matrix")
```
